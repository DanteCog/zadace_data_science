8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
#studenti.loc[studenti.EthnicGroup.isna(), 'EthnicGroup'] = studenti.loc[:, 'EthnicGroup'].mode()
studenti['EthnicGroup'].fillna(studenti['EthnicGroup'].mode()[0], inplace=True) # radi inplace = True Ne trebamo stavljati 'studenti[column] = '
studenti['ParentMaritalStatus'].fillna(studenti['ParentMaritalStatus'].mode()[0], inplace=True)
studenti['WklyStudyHours'].fillna(studenti['WklyStudyHours'].mode()[0], inplace=True)

studenti['ParentEduc'].fillna('no formal education', inplace=True)

studenti['TestPrep'].fillna('none', inplace=True)

studenti.dropna(subset=['PracticeSport', 'IsFirstChild'], inplace=True)

studenti['NrSiblings'].fillna(studenti['NrSiblings'].median(), inplace = True)

studenti['TransportMeans'].fillna('walking', inplace=True)
studenti
studenti.isna().sum()
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python

ohe = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False).set_output(transform='pandas')
le = LabelEncoder()
le.fit(studenti.loc[:,'EthnicGroup'])
studenti.loc[:,'EthnicGroup'] = le.transform(studenti.loc[:,'EthnicGroup'])
le.fit(studenti.loc[:,'ParentEduc'])
studenti.loc[:,'ParentEduc'] = le.transform(studenti.loc[:,'ParentEduc'])
le.fit(studenti.loc[:,'TestPrep'])
studenti.loc[:,'TestPrep'] = le.transform(studenti.loc[:,'TestPrep'])
le.fit(studenti.loc[:,'novi_first'])
studenti.loc[:,'novi_first'] = le.transform(studenti.loc[:,'novi_first'])
le.fit(studenti.loc[:,'TransportMeans'])
studenti.loc[:,'TransportMeans'] = le.transform(studenti.loc[:,'TransportMeans'])
le.fit(studenti.loc[:,'WklyStudyHours'])
studenti.loc[:,'WklyStudyHours'] = le.transform(studenti.loc[:,'WklyStudyHours'])
le.fit(studenti.loc[:,'PracticeSport'])
studenti.loc[:,'PracticeSport'] = le.transform(studenti.loc[:,'PracticeSport'])
le.fit(studenti.loc[:,'ParentMaritalStatus'])
studenti.loc[:,'ParentMaritalStatus'] = le.transform(studenti.loc[:,'ParentMaritalStatus'])

ohetransform = ohe.fit_transform(studenti[['Gender']])
ohetransform_2 = ohe.fit_transform(studenti[['LunchType']])
studenti = pd.concat([studenti, ohetransform], axis=1).drop(columns = ['Gender'])
studenti = pd.concat([studenti, ohetransform_2], axis=1).drop(columns = ['LunchType'])

#ohe.fit(studenti[['Gender']])
#studenti[['Gender']] = ohe.transform(studenti[['Gender']])
#ohe.fit(studenti[['LunchType']])
#studenti[['LunchType']] = ohe.transform(studenti[['LunchType']])
#studenti.loc[:,'EthnicGroup'] = le.transform(studenti.loc[:,'EthnicGroup'])
#studenti.loc[:,'ParentEduc'] = le.transform(studenti.loc[:,'ParentEduc'])
#studenti.loc[:,'TestPrep'] = le.transform(studenti.loc[:,'TestPrep'])
#studenti.loc[:,'ParentMaritalStatus'] = le.transform(studenti.loc[:,'ParentMaritalStatus'])
#studenti.loc[:,'IsFirstChild'] = le.transform(studenti.loc[:,'IsFirstChild'])
#studenti.loc[:,'TransportMeans'] = le.transform(studenti.loc[:,'TransportMeans'])
#studenti.loc[:,'WklyStudyHours'] = le.transform(studenti.loc[:,'WklyStudyHours'])
studenti.info()
studenti.head()
```
